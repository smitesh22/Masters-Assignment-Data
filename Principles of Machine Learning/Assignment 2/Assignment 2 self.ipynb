```python
import pandas as pd
```

```python
data = pd.read_table("wildfires.txt", delim_whitespace=True)
```

```python
data['fire'] = data['fire'].map({"yes":1, "no":0})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = data.corr()
sns.heatmap(corr, cmap="Blues",annot = True)
plt.savefig("Heatmap")
```
