## Introduction
Last week we looked statistical machine translation, where we modelled the word-level translation probabilities to in order to evaluate the likliehood of a translated sentence. The SMT model used the coocurrence frequencies of the source and target words over a corpora of aligned texts to build a translation-probability and develop a global distribution over sentence translations given a source text. SMT models are quite limited and have been superseded by many modern methods. In these lab we'll explore neural machine translation models (NMT) which have come to be the state-of-the-art for most translation tasks.
 
## Encoder-Decoder Models for NMT
In this lab we'll implement a simple version of the Encoder-Decoder introduced for NMT use cases by [Bahdanu et al 2014](https://arxiv.org/abs/1409.0473). The neural architecture consists of two parts, an encoder which projects sequential input from a source language into a lantent space and a decoder which decodde that encoded input into a probability distribution over the vocabulary of the target language.

## Exercise 1.1
Fill in the code below for the Language object which will be used to keep track of the vocabulary and encoding methods for each langauge. For both English and French, we will reserve the first three spots in our vocabulary for the special tokens PAD (paddinging), SOS (start of sentence), and EOS (end of sentence). 

```python
from nltk.tokenize import word_tokenize 
from typing import List 
import re 

class Langauge:
  def __init__(self, language: str):
    self.language = language                            # Name of the langauge
    self.word2index = {"PAD": 0, "SOS": 1, "EOS": 2}    # Maps each word in vocab to id
    self.index2word = {0: "PAD", 1: "SOS", 2: "EOS"}    # Reverse map of id to word in vocab
    self.word2count = {}                                # Count of each word in vocab
    self.n_words = len(self.index2word)                 # number of words in vocab

  def addSentence(self, sentence: str):
    """ 
    Given a sentence, lowercase is and remove any punctuation. Tokenize the
    sentence and for each word in the tokenized list call the addWord method.
    """
    text = sentence.lower()
    clean_text = re.sub(r'[^\w\s]', '', text).strip()
    for word in word_tokenize(clean_text):
      self.addWord(word)
  
  def addWord(self, word: str):
    """
    For each input word, check if it exists in the the word2index. If it does 
    not, add the word to the word2index and set the value to the current 
    vocabulary length. Update the index2word entry as well which maps the token 
    id to the word. Finaally update the vocabulary count (n_words).

    If the word is already in the vocabulary, udpate the count.
    """
    if word not in self.word2index:
      self.word2index[word] = self.n_words
      self.word2count[word] = 1
      self.index2word[self.n_words] = word
      self.n_words += 1
    else:
      self.word2count[word] += 1

  def encodeSentence(self, sentence: str) -> List[int]:
    """
    Given a sentence:
      1. Lower case it
      2. Remove all punctuation
      3. Prepend SOS and append EOS to it.
      4. Tokenize it and return the word ids for each word in the tokenized list. If a word
      does not exist in the vocab, skip over it. 

      Return a list of word ids. 
    """
    text = sentence.lower()
    clean_text = re.sub(r'[^\w\s]', '',text).strip()
    clean_text = "SOS " + clean_text + " EOS"
    return [self.word2index[word] for word in word_tokenize(clean_text) if word in self.word2index]

  def decodeIds(self, ids: list) -> List[str]:
    """
    Given a list of word ids, look the ids in the index2word and return a
    string representing the decoded sentence. 
    """
    return " ".join([self.index2word[tok] for tok in ids])
```

Once you implemented the Language class, create seperate objects for english and french. Using the those objects iterate over the `dataset` dataframe and build out the english and french vocabularies using `addSentence` method

```python
from tqdm.notebook import tqdm 

english = Langauge("english")
french = Langauge("french")

for _, row in tqdm(dataset.iterrows(), total=len(dataset)):
  english.addSentence(row["sourceString"])
  french.addSentence(row["targetString"])
print(f"Size of English vocab: {french.n_words}")
print(f"Size of French vocab: {english.n_words}")
```

Next we'll create the features for training. We use the pad_sequences method from the Keras library as it provides an easy way to pad a list of variable length sequences.

```python
import torch 
from tensorflow.keras.utils import pad_sequences
import pandas as pd

def encode_features(
    df: pd.DataFrame, 
    english: Langauge,
    french: Langauge,
    pad_token: int = 0,
    max_seq_length = 10
  ):

  source = []
  target = []

  for _, row in df.iterrows():
    source.append(english.encodeSentence(row["sourceString"]))
    target.append(french.encodeSentence(row["targetString"]))

  source = pad_sequences(
      source,
      maxlen=max_seq_length,
      padding="post",
      truncating = "post",
      value=pad_token
    )

  target = pad_sequences(
      target,
      maxlen=max_seq_length,
      padding="post",
      truncating = "post",
      value=pad_token
    )
  
  return source, target

train_source, train_target = encode_features(train, english, french)
val_source, val_target = encode_features(val, english, french)
test_source, test_target = encode_features(test, english, french)

print(f"Shapes of train source {train_source.shape}, and target {train_target.shape}")
```

## Part 2. Implementing the model
In this section we'll walk through the model implementation. The encoder-decoder model consists of two neural sub-modules: an encoder and decoder. Both will have similiar structure. The encoder will consist of an embedding layer and bi-directional GRU (gated reccurent unit [link text](https://en.wikipedia.org/wiki/Gated_recurrent_unit)). The GRU in our encoder will be bi-directional, meaning it will see the input twice, left to right and right to left in order learn the various sequential dependecies of the language useage. 

Below is an implementation of the Encoder code.

```python
import torch 
import torch.nn as nn 
import torch.nn.functional as F

class EncoderGRU(nn.Module):
    def __init__(
        self, 
        input_vocab_size,  # size of source vocabulary  
        hidden_dim,        # hidden dimension of embeddings
        encoder_hid_dim,   # gru hidden dim
        decoder_hid_dim,   # decoder hidden dim 
        dropout_prob = .5
      ):
      
        super().__init__()
        self.embedding = nn.Embedding(input_vocab_size, hidden_dim)
        self.rnn = nn.GRU(hidden_dim, encoder_hid_dim, bidirectional = True)
        self.fc = nn.Linear(encoder_hid_dim * 2, decoder_hid_dim)
        self.dropout = nn.Dropout(dropout_prob)
        
    def forward(self, src):
        
        #src = [src len, batch size]
        embedded = self.dropout(self.embedding(src))
        
        #embedded = [src len, batch size, emb dim]
        outputs, hidden = self.rnn(embedded)
                
        #outputs = [src len, batch size, hid dim * num directions]
        #hidden = [n layers * num directions, batch size, hid dim]        
        #hidden is stacked [forward_1, backward_1, forward_2, backward_2, ...]
        #outputs are always from the last layer
        
        #hidden [-2, :, : ] is the last of the forwards GRU
        #hidden [-1, :, : ] is the last of the backwards GRU
        
        #initial decoder hidden is final hidden state of the forwards and backwards 
        #  encoder RNNs fed through a linear layer
        hidden = torch.tanh(self.fc(torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim = 1)))        
        return outputs, hidden
```

The decoder module is similiar to the encoder with the primary difference being the addition of the Attention layer. The attention layer takes in the previous hidden state of the decoder and all the forward and backward hidden states of the encoder. It then computes an attention vector over the length of the sentence which is used by the model to learn which part of the source sentence is most salient when translating the current token. Below is an implementation of the attention layer.

```python
class Attention(nn.Module):
    def __init__(
        self, 
        enc_hid_dim,      # Encoder hidden dimension
        dec_hid_dim       # Decoder hidden dimension 
      ):
        super().__init__()
        
        self.attn = nn.Linear((enc_hid_dim * 2) + dec_hid_dim, dec_hid_dim)
        self.v = nn.Linear(dec_hid_dim, 1, bias = False)
        
    def forward(self, hidden, encoder_outputs):
        
        #hidden = [batch size, dec hid dim]
        #encoder_outputs = [src len, batch size, enc hid dim * 2]
        batch_size = encoder_outputs.shape[1]
        src_len = encoder_outputs.shape[0]
        
        #repeat decoder hidden state src_len times
        hidden = hidden.unsqueeze(1).repeat(1, src_len, 1)
        
        encoder_outputs = encoder_outputs.permute(1, 0, 2)
        #hidden = [batch size, src len, dec hid dim]
        #encoder_outputs = [batch size, src len, enc hid dim * 2]
        
        energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim = 2))) 
        
        #energy = [batch size, src len, dec hid dim]
        attention = self.v(energy).squeeze(2)
        
        #attention output: [batch size, src len]
        return F.softmax(attention, dim=1)
```

Next we implement the Decoder layer. The decoder layer contains an attention layer, seperate embedding layer over the target language vocabulary, a GRU layer and finally an output layer mapping the decoder latent representation for each word to a probability distribution over the target vocabulary. Below we implement the decoder layer.

```python
class DecoderGRU(nn.Module):
    def __init__(
        self, 
        target_vocab_size,    # Size of target vocab 
        hidden_dim,           # hidden size of embedding  
        enc_hid_dim, 
        dec_hid_dim, 
        dropout
      ):
        super().__init__()

        self.output_dim = target_vocab_size
        self.attention = Attention(enc_hid_dim, dec_hid_dim)
        
        self.embedding = nn.Embedding(target_vocab_size, hidden_dim)
        
        self.rnn = nn.GRU((enc_hid_dim * 2) + hidden_dim, dec_hid_dim)
        
        self.fc_out = nn.Linear(
            (enc_hid_dim * 2) + dec_hid_dim + hidden_dim, 
            target_vocab_size
          )
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, input, hidden, encoder_outputs):
             
        #input = [batch size]
        #hidden = [batch size, dec hid dim]
        #encoder_outputs = [src len, batch size, enc hid dim * 2]
        
        input = input.unsqueeze(0)  # [1, batch size]
        
        embedded = self.dropout(self.embedding(input))  # [1, batch size, emb dim]
        
        a = self.attention(hidden, encoder_outputs)     # [batch size, src len]
        a = a.unsqueeze(1)                              # [batch size, 1, src len]
        
        encoder_outputs = encoder_outputs.permute(1, 0, 2) # [batch size, src len, enc hid dim * 2]
        
        weighted = torch.bmm(a, encoder_outputs)           # [batch size, 1, enc hid dim * 2]
        weighted = weighted.permute(1, 0, 2)               # [1, batch size, enc hid dim * 2]
        
        rnn_input = torch.cat((embedded, weighted), dim = 2) # [1, batch size, (enc hid dim * 2) + emb dim]

        
        #output = [seq len, batch size, dec hid dim * n directions]
        #hidden = [n layers * n directions, batch size, dec hid dim]    
        output, hidden = self.rnn(rnn_input, hidden.unsqueeze(0))
        
        #seq len, n layers and n directions will always be 1 in this decoder, therefore:
        #output = [1, batch size, dec hid dim]
        #hidden = [1, batch size, dec hid dim]
        
        embedded = embedded.squeeze(0)
        output = output.squeeze(0)
        weighted = weighted.squeeze(0)
        
        prediction = self.fc_out(torch.cat((output, weighted, embedded), dim = 1)) # [batch size, output dim]
        return prediction, hidden.squeeze(0)
```

Finally we implement the Encoder Decoder Model which is a Seq2Seq model. The seq2seq model facilitates the passing of the encoder hidden states to the decoder. Teacher forcing is a method to ensure the decoder sees the correct output during training.

```python
import random 
class EncoderDecoder(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        
        self.encoder = encoder
        self.decoder = decoder
        
    def forward(self, src, trg, teacher_forcing_ratio = 0.5):
        
        #src = [src len, batch size]
        #trg = [trg len, batch size]
        #teacher_forcing_ratio is probability to use teacher forcing
        #e.g. if teacher_forcing_ratio is 0.75 we use teacher forcing 75% of the time     
        batch_size = src.shape[1]
        trg_len = trg.shape[0]
        trg_vocab_size = self.decoder.output_dim
        
        #tensor to store decoder outputs
        outputs = torch.zeros(trg_len, batch_size, trg_vocab_size)
        
        #encoder_outputs is all hidden states of the input sequence, back and forwards
        #hidden is the final forward and backward hidden states, passed through a linear layer
        encoder_outputs, hidden = self.encoder(src)
                
        #first input to the decoder is the <sos> tokens
        input = trg[0,:]
        
        for t in range(1, trg_len):     
            #insert input token embedding, previous hidden state and all encoder hidden states
            #receive output tensor (predictions) and new hidden state
            output, hidden = self.decoder(input, hidden, encoder_outputs)
            
            #place predictions in a tensor holding predictions for each token
            outputs[t] = output
            
            #decide if we are going to use teacher forcing or not
            teacher_force = random.random() < teacher_forcing_ratio
            
            #get the highest predicted token from our predictions
            top1 = output.argmax(1) 
            
            #if teacher forcing, use actual next token as next input
            #if not, use predicted token
            input = trg[t] if teacher_force else top1
        return outputs
```

```python
INPUT_DIM = english.n_words
OUTPUT_DIM = french.n_words
ENC_EMB_DIM = 256
DEC_EMB_DIM = 256
ENC_HID_DIM = 128
DEC_HID_DIM = 128
ENC_DROPOUT = 0.5
DEC_DROPOUT = 0.5

enc = EncoderGRU(INPUT_DIM, ENC_EMB_DIM, ENC_HID_DIM, DEC_HID_DIM, ENC_DROPOUT)
dec = DecoderGRU(OUTPUT_DIM, DEC_EMB_DIM, ENC_HID_DIM, DEC_HID_DIM, DEC_DROPOUT)

model = EncoderDecoder(enc, dec)

def init_weights(m):
    for name, param in m.named_parameters():
        if 'weight' in name:
            nn.init.normal_(param.data, mean=0, std=0.01)
        else:
            nn.init.constant_(param.data, 0)
            
model.apply(init_weights)
```

Finally we can implement our training loop below. 

```python
from tqdm.notebook import tqdm
import numpy as np 
optimizer = torch.optim.Adam(model.parameters())

device = "cuda:0" if torch.cuda.is_available() else "cpu"

model.to(device)

EPOCHS = 1
best_val_loss = float('inf')

for epoch in range(EPOCHS):

  model.train()
  epoch_loss = 0
  for batch in tqdm(train_dl, total=len(train_dl)):

     src = batch[0].transpose(1, 0).to(device)
     trg = batch[1].transpose(1, 0).to(device)

     optimizer.zero_grad()

     output = model(src, trg)

     output_dim = output.shape[-1]
     output = output[1:].view(-1, output_dim).to(device)
     trg = trg[1:].reshape(-1)
     
     loss = F.cross_entropy(output, trg)
     loss.backward()

     torch.nn.utils.clip_grad_norm_(model.parameters(), 1)
     optimizer.step()
     epoch_loss += loss.item()

  train_loss = round(epoch_loss / len(train_dl), 3)
  
  eval_loss = 0
  model.eval()
  for batch in tqdm(val_dl, total=len(val_dl)):
    src = batch[0].transpose(1, 0).to(device)
    trg = batch[1].transpose(1, 0).to(device)

    with torch.no_grad():
      output = model(src, trg)
      
      output_dim = output.shape[-1]
      output = output[1:].view(-1, output_dim).to(device)
      trg = trg[1:].reshape(-1)
      
      loss = F.cross_entropy(output, trg)
      
      eval_loss += loss.item()
  
  val_loss = round(eval_loss / len(val_dl), 3)
  print(f"Epoch {epoch} | train loss {train_loss} | train ppl {np.exp(train_loss)} | val ppl {np.exp(val_loss)}")


  if val_loss < best_val_loss:
    best_val_loss = val_loss
    torch.save(model.state_dict(), 'best-model.pt')  
  
```

## Evaluation and Inference
In this section we'll implement a method to generate inferences from our Seq2Seq model.

```python
def translate_sentence(
    text: str, 
    model: EncoderDecoder, 
    english: Langauge,
    french: Langauge,
    device: str,
    max_len: int = 10,
  ) -> str:

  # Encode english sentence and convert to tensor
  input_ids = english.encodeSentence(text)
  input_tensor = torch.LongTensor(input_ids).unsqueeze(1).to(device)

  # Get encooder hidden states
  with torch.no_grad():
    encoder_outputs, hidden = model.encoder(input_tensor)

  # Build target holder list
  trg_indexes = [french.word2index["SOS"]]

  # Loop over sequence length of target sentence
  for i in range(max_len):
    trg_tensor = torch.LongTensor([trg_indexes[-1]]).to(device)
    
    # Decode the encoder outputs with respect to current target word
    with torch.no_grad():
      output, hidden = model.decoder(trg_tensor, hidden, encoder_outputs)
    
    # Retrieve most likely word over target distribution
    pred_token = torch.argmax(output).item()
    trg_indexes.append(pred_token)

    if pred_token == french.word2index["EOS"]:
      break

  return "".join(french.decodeIds(trg_indexes))
```
